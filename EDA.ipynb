```python
## Univariate analysis
```

```python
credit['LIMIT_BAL'].value_counts(sort=True).plot.bar(title="Freq dist of Limits")
```

```python
credit['AGE'].value_counts(sort=True).plot.bar(title="Freq Age")
```

```python
df = credit
df['EDUCATION'] = df.apply(lambda x: 4 if x['EDUCATION'] >= 5 else x['EDUCATION'],axis=1)
df = df[df.MARRIAGE != 0]
df.rename(columns={'PAY_0':'PAY_1'}, inplace=True)
df.rename(columns={'default payment next month':'default_payment_next_month'}, inplace=True)
col_names = ['LIMIT_BAL','SEX', 'MARRIAGE', 'EDUCATION', 'AGE', 'BILL_AMT1']


fig, ax = plt.subplots(len(col_names), figsize=(16,12))

for i, col_val in enumerate(col_names):

    sns.distplot(df[col_val], hist=True, ax=ax[i])
    ax[i].set_title('Freq dist '+col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)
    ax[i].set_ylabel('Count', fontsize=8)

plt.show()
```

```python
### LIMIT_BAL is left-skewed
### marriage is a candidate to dump
### BIL_AMT1 and AGE are left-skewed
```

```python
## Pair wise dating/scatter plot
df.drop(['ID'], axis=1)
sns.pairplot(df)
```

```python
## We can decipher that some variables like sex and marriage have little influence
## but no clear increasing decreasing patterns observed in the pair plot
## There seems to be some positive correlation between PAY_1 and PAY_3 but thats about it
```

```python
## Target value imbalance
temp = df["default_payment_next_month"].value_counts()
df1 = pd.DataFrame({'default_payment_next_month': temp.index,'values': temp.values})
plt.figure(figsize = (3,3))
plt.title('Default Credit Card Clients\n (Default = 0, Not Default = 1)')
sns.set_color_codes("pastel")
sns.barplot(x = 'default_payment_next_month', y="values", data=df1)
locs, labels = plt.xticks()
plt.show()
## Around 6500 out of 30000 seend to be defaulting next month, no data imbalance w.r.t target variable
```

```python
## Amount of credit limit grouped by default payment next month

## CREATE THE TWO CLASSES
class_0 = df.loc[df['default_payment_next_month'] == 0]["LIMIT_BAL"]
class_1 = df.loc[df['default_payment_next_month'] == 1]["LIMIT_BAL"]
plt.figure(figsize = (14,6))
plt.title('Default amount of credit limit  - grouped by Payment Next Month (Density Plot)')
sns.set_color_codes("pastel")
sns.distplot(class_1,kde=True,bins=200, color="red")
sns.distplot(class_0,kde=True,bins=200, color="blue")
plt.show()

## The class of defaulter are coming more on the lower end of the spectrum of LIMITS
```

```python
### Most of defaults are for credit limits 0-100,000 
## (and density for this interval is larger for defaults than for non-defaults). 

## Credit limit vs SEX
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))
s = sns.boxplot(ax = ax1, x="SEX", y="LIMIT_BAL", hue="SEX",data=df, palette="PRGn",showfliers=True)
s = sns.boxplot(ax = ax2, x="SEX", y="LIMIT_BAL", hue="SEX",data=df, palette="PRGn",showfliers=False)
plt.show();
```

```python
## Ok Sex is not a factor, except for an outlier one for the Female sex's

## Amount of BILL_AMT1 grouped by default payment next month
## CREATE THE TWO CLASSES
class_0 = df.loc[df['default_payment_next_month'] == 0]["BILL_AMT1"]
class_1 = df.loc[df['default_payment_next_month'] == 1]["BILL_AMT1"]
plt.figure(figsize = (14,6))
plt.title('BILL_AMT1  - grouped by Payment Next Month (Density Plot)')
sns.set_color_codes("pastel")
sns.distplot(class_1,kde=True,bins=200, color="red")
sns.distplot(class_0,kde=True,bins=200, color="blue")
plt.show()
```

```python
## Clerarly the ones receiving credit for that month are not
## defaulting the next month most likely
## Amount of BILL_AMT6 grouped by default payment next month
## CREATE THE TWO CLASSES
class_0 = df.loc[df['default_payment_next_month'] == 0]["BILL_AMT6"]
class_1 = df.loc[df['default_payment_next_month'] == 1]["BILL_AMT6"]
plt.figure(figsize = (14,6))
plt.title('BILL_AMT6  - grouped by Payment Next Month (Density Plot)')
sns.set_color_codes("pastel")
sns.distplot(class_1,kde=True,bins=200, color="red")
sns.distplot(class_0,kde=True,bins=200, color="green")
plt.show()

## Even for BILL_AMT6 pattern is same
```
